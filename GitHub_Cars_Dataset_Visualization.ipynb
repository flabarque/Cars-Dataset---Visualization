# **Cars Dataset - Visualization**

The period between 1970 and 1982 marked a significant shift in the United States car industry. American production shifted from heavy, powerful, six- and eight-cylinder cars with poor gas mileage to lighter, less powerful, four-cylinder cars with higher fuel efficiency.

The global auto industry, including Americans and their European and Japanese competitors, raised overall miles per gallon (MPG) by focusing on four-cylinder cars and making them more fuel-efficient.

---

The dataset contains 398 unique cars along with their respective specification, origin country, and model year. This dataset is a version of the CRCARS dataset of Donoho & Ramos (1982): "PRIMDATA: Data Sets for Use With PRIM-H", used at the 1983 American Statistical Association Exposition. The StatLib library at Carnegie Mellon University stores the original version of this dataset (http://lib.stat.cmu.edu/datasets/cars.data). However, a "friendly" *.cvs version (Comma-Separated Value (CSV) files) is available at the following link:

**Dataset in CSV:** https://data.world/dataman-udit/cars-data (download Cars_Multi.cvs)

**Attributes:**

*   id: string (Unique values - PK)
*   mpg: continuous
*   cylinders: multi-valued discrete
*   displacement: continuous
*   horsepower: continuous
*   weight: continuous
*   acceleration: continuous
*   model: multi-valued discrete
*   origin: multi-valued discrete (FK)
*   car name: string (Unique values)

**Definitions:**

*   mpg: also known as miles per gallon, is a measurement of fuel efficiency.
*   displacement: engine displacement determines the horsepower and torque that an engine produces, as well as how much fuel that engine consumes. Higher displacement equals more power, and lower displacement equals more efficiency. 
*   horsepower: is used to explain the power of the car’s engine concerning its weight. Heavier cars will require more horsepower to move.
*   acceleration: is calculated when the car is not in motion (0 mph), until the amount of time it takes to reach a velocity of 60 miles per hour.

Please, note that Pandas' DataFrame.histogram() uses matplotlib.pyplot.hist() to draw a Matplotlib histogram of the input DataFrame using subplots, e.g., <matplotlib.axes._subplots.AxesSubplot object at 0x7f1ccaccbd50>. Subplots mean a group of smaller axes (where each axis is a plot) that can exist together within a single figure. In other word, the method provides a way to plot multiple plots on a single figure.

We can create a function to plot multiple plots (subplots) into a figure using Matplotlib. This will provide us the freedom, not only to plot multiple histograms in a figure, but any other type of graphical representation. In addition, it will allow us to edit many more figure parameters as we are using the Matplotlib library.

```python
#creating a function to built several histograms into a single figure of the DataFrame from Matplotlib

#creating and defining funtion parameters
#dataframe: our data 
#features: our targets (i.e., which values we want to plot)
#rows and cols: number of rows and columns with subplots within figure   
def histograms_plot(dataframe, features, rows, cols):
  #defining figure parameters
  fig=plt.figure(figsize=(24,10))
  #for loop to implement subplots for each feature
  for i, feature in enumerate(features):
      #adding subplots to figure
      ax=fig.add_subplot(rows,cols,i+1)
      #calling histograms for each subplot
      dataframe[feature].hist(bins=20,ax=ax)
      #adding title to each subplot
      ax.set_title(feature+' distribution', fontsize=15, fontweight='bold')
  #adjusting subplot parameters 
  fig.tight_layout()  
  #showing figure
  plt.show()
```

However, we need to define a list with the numerical values (features). Otherwise, the function will call all the features including *origin* and *name*, that are not numerical data. 

```python
#defining a list with the numerical values
num_val = ['mpg', 'cylinders','displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
```

```python
#histograms for each numeric value of the DataFrame from Matplotlib
histograms_plot(df_drop,num_val,3,3)
```

**Sources:**
*   https://stackoverflow.com/questions/50773877/create-for-loop-to-plot-histograms-for-individual-columns-of-dataframe-with-seab

Additionally, we can use the Seaborn library to plot each histogram with a kernel density estimator (i.e., kde). The kde is a probability density function that works by plotting out the data and creating a curve of its distribution. That curve has positive values (including zero), and the area below it should be equal to one. The kde gives a much smoother result, while a histogram produces more steep edges (i.e., fixed area rectangles).

```python
#creating a function to built several histograms with their kde into a single figure of the DataFrame from Seaborn

#creating and defining funtion parameters
#dataframe: our data 
#features: our targets (i.e., which values we want to plot)
#rows and cols: number of rows and columns with subplots within figure   
def histograms_kde_plot(dataframe, features, rows, cols):
  #defining figure parameters
  fig=plt.figure(figsize=(24,8))
  #for loop to implement subplots for each feature
  for i, feature in enumerate(features):
      #adding subplots to figure
      ax=fig.add_subplot(rows,cols,i+1)
      #calling histograms for each subplot with their kde
      sns.histplot(x= dataframe[feature],bins=20,ax=ax,kde=True)
      #adding title to each subplot
      ax.set_title(feature+' distribution', fontsize=15, fontweight='bold')
  #adjusting subplot parameters 
  fig.tight_layout()  
  #showing figure
  plt.show()
```

Standard kde assumes and estimates a density function of a continuous random variable. For this reason, we need to define a new list with only the continuous numerical values (features).

```python
#defining a list with the continue (i.e., not discrete) numerical values
con_num_val = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
```

```python
#histograms for each numeric value of the DataFrame from Matplotlib
histograms_kde_plot(df_drop,con_num_val,2,3)
```

### **Boxplot**

A boxplot is a method for graphically depicting groups of numerical data through their quartiles (i.e., “minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). In other words, a boxplot is a graph that gives you a good indication of how the values in the data are spread out.

**Sources:**
*   https://en.wikipedia.org/wiki/Histogram
*   https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51

```python
#creating a function to built several boxplots into a single figure of the DataFrame from Seaborn

#creating and defining funtion parameters
#dataframe: our data 
#features: our targets (i.e., which values we want to plot)
#rows and cols: number of rows and columns with subplots within figure   
def boxplots_plot(dataframe, features, rows, cols):
  #defining figure parameters
  fig=plt.figure(figsize=(24,10))
  #for loop to implement subplots for each feature
  for i, feature in enumerate(features):
      #adding subplots to figure
      ax=fig.add_subplot(rows,cols,i+1)
      #calling boxplot for each subplot
      sns.boxplot(y= dataframe[feature],ax=ax)
      #adding title to each subplot
      ax.set_title(feature+' boxplot', fontsize=15, fontweight='bold')
  #adjusting subplot parameters 
  fig.tight_layout()  
  #showing figure
  plt.show() 
```

```python
#boxplots for each numeric value of the DataFrame from Seaborn
boxplots_plot(df_drop,con_num_val,1,5)
```

The box plots show that *horsepower* and *acceleration* have outliers (i.e., show as black diamonds). Outliers increase the error variance and reduce the power of statistical tests. They can cause bias and influence estimates. Thus, investigating outliers is crucial in data cleaning (not the scope of this presentation).

### **Scatter Plot and Linear Regression**

A scatter plot is a graph designed to show the relationship between two variables using dots to represent their values. If the dots show an uphill pattern from left to right, it will indicate a positive correlation between the variables. A simple linear regression will model the relationship between those variables by fitting a linear equation to the observed data. 

### **For example: How weight affects MPG?**



```python
#defining figure parameters
plt.figure(figsize=(7, 4))
#scatter plot of weight vs MPG values of the DataFrame from Seaborn
sns.regplot(x='mpg', y='weight', data=df_drop, fit_reg=False, scatter_kws={'alpha': 0.2})
#showing figure
plt.show()

#fit_reg=False --> no simple linear regression
#scatter_kws={'alpha': 0.2} --> dots transparency
```

The scatter plot shows the negative correlation between *weight* and *MPG*, which means that heavier cars have lower efficiency (i.e., lower MPG). 

```python
#defining figure parameters
plt.figure(figsize=(7, 4))
#scatter plot of weight vs MPG values with linear regression of the DataFrame from Seaborn
sns.regplot(x='mpg', y='weight', data=df_drop, scatter_kws={'alpha': 0.2})
#showing figure
plt.show()

#scatter_kws={'alpha': 0.2} --> dots transparency
```

The line above the scatter plot shows that the negative correlation between *weight* and *MPG* fits a simple linear regression.

```python
#joint plot of weight vs MPG values with linear regression of the DataFrame from Seaborn
sns.jointplot(x='mpg', y='weight', data=df_drop, kind='reg', height=6.5)
#showing figure
plt.show()
```

A Joint Plot quickly visualizes and analyzes the relationship between two variables and describes their distributions on the same plot.

**Sources:**

*  https://www.hackerearth.com/blog/developers/data-visualization-techniques/

### **How do all the other continuous numerical features afect *MPG*?**

We can adapt our previous function to plot multiple regression plots (subplots) for each feature against any other into a figure using Matplotlib. 

```python
#creating a function to built several regplots for each feature against MPG into a single figure of the DataFrame from Seaborn

#creating and defining funtion parameters
#dataframe: our data 
#features: our targets (i.e., which values we want to plot)
#rows and cols: number of rows and columns with subplots within figure   
def regplots_plot(dataframe, features, rows, cols):
  #defining figure parameters
  fig=plt.figure(figsize=(20,10))
  #for loop to implement subplots for each feature
  for i, feature in enumerate(features):
      #adding subplots to figure
      ax=fig.add_subplot(rows,cols,i+1)
      #calling regplot for each subplot 
      sns.regplot(x='mpg', y=dataframe[feature], data=df_drop, scatter_kws={'alpha': 0.2},ax=ax)
      #adding title to each subplot
      ax.set_title(feature+' linear regression', fontsize=12, fontweight='bold')
  #adjusting subplot parameters 
  fig.tight_layout()  
  #showing figure
  plt.show() 
```

In this scenario, we need to define a new list with the continuous numerical values (features) except MPG. Otherwise, we will be checking how MPG affects itself.

```python
#defining a list with the continue numerical values except MPG
con_num_val_no_mpg = ['displacement', 'horsepower', 'weight', 'acceleration']
```

```python
#regplots for each numeric value of the DataFrame from Seaborn
regplots_plot(df_drop,con_num_val_no_mpg,3,3)
```

The scatter plots show the negative correlation and simple linear regression fit between *displacement* and *MPG*, *horsepower* and *MPG*, and *weight* and *MPG*. In other words, high-displacement engines, high power engines, and heavier cars have lower efficiency (i.e., lower MPG), respectively. In addition, note the positive correlation and simple linear regression fit between *acceleration* and *MPG*. This means that cars with high acceleration have higher efficiency.
